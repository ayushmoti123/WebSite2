```python
import pandas as pd
```

```python
import numpy as np
```

```python
#preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#splitting

from sklearn.model_selection import train_test_split

#Model
from sklearn.linear_model import LinearRegression

#Pipeline
from sklearn.pipeline import Pipeline

#Metrics

from sklearn.metrics import mean_squared_error,r2_score


```

```python
#model
from sklearn.tree import DecisionTreeRegressor
```

```python
#better_model
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
```

```python
data=pd.read_csv("listings.csv")
#data.head()
#data["neighbourhood_group"].isna().count()
#data["license"].isna().sum()
#(data["number_of_reviews_ltm"]==0).sum()
#data.isna().sum()
```

```python
data_copy=data.copy()
#data_copy.drop("neighbourhood_group",axis=1,inplace=True)
#data_copy.drop("license",axis=1,inplace=True)
#data_copy.isna().sum()
#data_copy["price","minimum_nights"
```

```python
features=['room_type','minimum_nights','calculated_host_listings_count','availability_365']
data_baseline=data_copy.dropna(subset=features+['price'])
```

```python
from sklearn import preprocessing
X=data_baseline[features]
Y=data_baseline['price']
preprocesser=ColumnTransformer(
    [('cat',OneHotEncoder(handle_unknown='ignore'),['room_type']),
    ('num','passthrough',['minimum_nights','calculated_host_listings_count','availability_365'])
    ])

```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model_lr=Pipeline([('preprocess',preprocesser),('model',LinearRegression())])
model_lr.fit(x_train,y_train)
pred_lr=model_lr.predict(x_test)
```

```python
rmse_lr=np.sqrt(mean_squared_error(y_test,pred_lr))
print(rmse_lr)
```

```python
model_dt=Pipeline([('preprocess',preprocesser),('model',DecisionTreeRegressor(random_state=42))])
model_dt.fit(x_train,y_train)
pred_dt=model_dt.predict(x_test)
```

```python
rmse_dt=np.sqrt(mean_squared_error(y_test,pred_dt))
print(rmse_dt)
```

```python
model_rf=Pipeline([('preprocess',preprocesser),('model',RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    max_depth=None,
    n_jobs=-1
))])
model_rf.fit(x_train,y_train)
pred_rf=model_rf.predict(x_test)
```

```python
rmse_rf=np.sqrt(mean_squared_error(y_test,pred_rf))
print(rmse_rf)
```

```python
model_gb=Pipeline([('preprocess',preprocesser),('model',GradientBoostingRegressor(
    n_estimators=300,
    random_state=42,
    max_depth=4,
    learning_rate=0.05
))])
model_gb.fit(x_train,y_train)
pred_gb=model_gb.predict(x_test)
```

```python
rmse_gb=np.sqrt(mean_squared_error(y_test,pred_gb))
print(rmse_gb)
```
